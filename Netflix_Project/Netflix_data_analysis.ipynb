# Netflix Movies and TV Shows Data Analysis

## Checking Missing Values :
This step checks the number of missing values present in each column.

This step checks whether duplicate rows are present in the dataset.

## Handling Missing Values :
Missing values are removed to make the dataset clean for analysis.

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

## Selecting Important Columns :
Only relevant columns are selected for focused analysis.

```python
df[['type', 'release_year', 'rating', 'country']]
```

## Feature Engineering :
New features are created from existing columns to improve analysis.

```python
df['Content_Age'] = np.where(df['release_year'] < 2015, 'Old Content', 'New Content')
```

```python
df[['release_year', 'Content_Age']].head()
```

## Categorical Analysis :
This step analyzes categorical columns to understand content distribution.

```python
df['type'].value_counts()
```

```python
df['rating'].value_counts()
```

## Content Type Analysis :
This analysis shows the count of Movies and TV Shows available on Netflix.

```python
df.groupby('type').size()
```

## Content Age Analysis :
This analysis shows the distribution of old and new content on Netflix.

```python
df['Content_Age'].value_counts()
```

## Visualization: Movies vs TV Shows 
A bar chart is used to visualize the count of Movies and TV Shows on Netflix.

```python
df['type'].value_counts().plot(kind='bar', color='red')
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

## Visualization: Content by Rating :
This bar chart shows the distribution of Netflix content based on ratings.

```python
df['rating'].value_counts().plot(kind='bar', color='red')
plt.title('Netflix Content by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
