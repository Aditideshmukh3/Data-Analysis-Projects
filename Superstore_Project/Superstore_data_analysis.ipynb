# Global Superstore Sales Data Analysis

## Objective :
The objective of this project is to analyze global sales data to understand sales performance,
customer behavior, and product trends using Python, NumPy, and Pandas.
Basic data analysis, data cleaning, feature engineering, and visualization techniques are applied.

## Libraries Used :
- Pandas – for data manipulation and analysis  
- NumPy – for numerical operations  
- Matplotlib – for data visualization  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Dataset Loading :
In this step, the Global Superstore dataset is loaded from an Excel file using Pandas.

```python
df = pd.read_excel(r"D:\global_superstore_2016.xlsx")
```

## Viewing the Dataset :
The head() function is used to display the first five rows of the dataset.

```python
df.head(5)
```

## Dataset Understanding :
Basic functions are used to understand the structure and size of the dataset.

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

## Checking Missing Values :
This step checks the number of missing (NaN) values present in each column of the dataset.

```python
df.isnull().sum()
```

## Dropping Missing Values :
Rows containing missing values are removed to make the dataset clean for analysis.

```python
df_clean = df.dropna()
```

```python
df_clean.isnull().sum()
```

## Selecting Data Using iloc :
The iloc function is used to select rows and columns based on index positions.

```python
df_clean.iloc[0:5, 0:6]
```

## Selecting Data Using Condition :
This step selects data based on a specific condition.

```python
df_clean[df_clean['Sales'] > 500]
```

## Grouping Data :
The groupby function is used to group data based on a column and perform analysis.

```python
df_clean.groupby('Region')['Sales'].sum()
```

## Feature Engineering :
A new feature is created to categorize sales as High or Low.

```python
df_clean.loc[:, 'Sales_Category'] = np.where(
    df_clean['Sales'] > 400, 'High Sales', 'Low Sales'
)
```

```python
df_clean[['Sales', 'Sales_Category']].head()
```

## Handling Missing Values Using fillna() :
The fillna() function is used to replace missing values with a specific value.

```python
df_fill = df.copy()
df_fill['Discount'] = df_fill['Discount'].fillna(0)
```

```python
df_fill['Discount'].isnull().sum()
```

## Concatenating DataFrames :
The concat() function is used to combine two DataFrames vertically.

```python
df_part1 = df_clean.head(5)
df_part2 = df_clean.tail(5)
```

```python
df_concat = pd.concat([df_part1, df_part2])
df_concat
```

## Merge Operation :
The merge() function is used to combine DataFrames based on a common column.

```python
sales_df = df_clean[['Order ID', 'Sales']]
profit_df = df_clean[['Order ID', 'Profit']]
```

```python
df_merged = pd.merge(sales_df, profit_df, on='Order ID')
df_merged.head()
```

## Join Operation :
The join() function is used to combine DataFrames using index.

```python
sales_df_indexed = sales_df.set_index('Order ID')
profit_df_indexed = profit_df.set_index('Order ID')
```

```python
df_joined = sales_df_indexed.join(profit_df_indexed)
df_joined.head()
```

## Visualization- Sales by Region :
A bar chart is used to visualize total sales across regions.

```python
df_clean.groupby('Region')['Sales'].sum().plot(kind='bar', color='purple')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

Various data handling techniques such as missing value handling, data selection, grouping, feature engineering,
concatenation, merge, and join operations were performed.
Basic visualization helped in understanding regional sales performance.
