# Titanic Data Analysis

**Objective:** 
This project focuses on analyzing the Titanic passenger dataset to understand factors affecting survival. This project includes data exploration, cleaning, feature engineering, and insights extraction using Python (pandas, NumPy).  

## Libraries Used :

- **pandas**: used for loading, exploring, and manipulating tabular data  
- **NumPy**: used for numerical calculations such as mean and median  
- **matplotlib**: used for basic data visualization such as bar charts and histograms

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Loading the Dataset :

The dataset is loaded using the `read_csv()` function from pandas. This function reads a CSV file and converts it into a DataFrame, which is a table-like data structure with rows and columns.

```python
df = pd.read_csv("D:\Titanic Data.csv")
```

## Previewing the Dataset :

The `head()` function displays the first five rows of the dataset.
This helps in understanding the structure of the data,
column names, and the type of values stored in each column.

```python
df.head()
```

## Dataset Shape :

The `shape` attribute is used to find the number of rows and columns
present in the dataset. This gives an idea about the size of the data.

```python
df.shape
```

## Dataset Information :

The `info()` function provides a concise summary of the dataset.
It displays column names, data types, number of non-null values,
and helps identify missing data and incorrect data types.

```python
df.info()
```

```python
df.describe()
```

## Checking Data Types :

This step is used to understand the data type of each column.
It helps identify numerical and categorical variables.

```python
df.dtypes
```

## Accessing a Single Column :

This step demonstrates how to select a single column from the DataFrame.
Here, the Age column is accessed.

```python
df['Age']
```

## Accessing Multiple Columns :

This step is used to select more than one column from the dataset.

```python
df[['Age', 'Sex', 'Survived']]
```

## Checking Missing Values :

This function is used to identify missing values in each column of the dataset.
It helps verify the quality of the data.

```python
df.isnull().sum()
```

## Checking Duplicate Rows :

This step checks whether duplicate rows are present in the dataset.
Duplicate records can affect analysis accuracy.

```python
df.duplicated().sum()
```

## Sorting the Dataset :
 
This step sorts the dataset based on Age in ascending order.
Sorting helps in better understanding of data distribution.

```python
df.sort_values(by='Age')
```

## Conditional Filtering :

This step filters the dataset to show passengers whose age is greater than 30.
It demonstrates conditional selection of data.

```python
df[df['Age'] > 30]
```

## Using loc for Data Selection :

The loc function is used to access rows and columns using labels.

```python
df.loc[0:5, ['Age', 'Sex', 'Survived']]
```

## Using iloc for Data Selection :

The iloc function is used to access rows and columns using index positions.

```python
df.iloc[0:5, 0:4]
```

## GroupBy Analysis :

This step groups the dataset based on survival status
and calculates the average age of passengers.

```python
df.groupby('Survived')['Age'].mean()
```

## Categorical Analysis :

This step analyzes survival count based on gender.

```python
df.groupby('Sex')['Survived'].sum()
```

## Survival Count Visualization :

This bar chart shows the number of passengers who survived
and those who did not survive.

```python
plt.bar(df['Survived'].value_counts().index, df['Survived'].value_counts())
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Count')
plt.show()
```

## Age Distribution :

This histogram shows the distribution of passengers based on age.

```python
plt.hist(df['Age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```
